# 1 . Total 10K companies

```python
#all companies
import requests
import json

url = "http://44.193.54.227:8000/backend/get_index_documents"

payload = json.dumps({
  "elastic_indx": [
    "all"
  ]
})
headers = {
  'Content-Type': 'application/json',
  'Authorization': 'Token 26f9b124da0da589be592f666da4f45c3981f355'
}

response = requests.request("POST", url, headers=headers, data=payload)

companies = response.json()
```

```python
def total_10K():
    total_companies = []
    for i in range(len(companies['data']['10k'])):
        #for getting unique id , replace PDF with id
        total_companies.append(companies['data']['10k'][i]['PDF'])
    return len(total_companies),total_companies
```

```python
total_10K()
```

# 2. Getting all unique ids per companies

```python
#id per company
import requests
import json

url = "http://44.193.54.227:8000/backend/sql_query"

payload = json.dumps({
  "elastic_indx": "10k",
  "select_fields": [
    "id",
    "PDF",
    "table_xml",
    "table_xls",
    "table_html",
    "page",
    "table_thumbnail_png"
  ],
  "type": [
    "table"
  ],
  "documents_set": "test",
  "group_by": [
    "PDF"
  ],
  "max_results": 200
})
headers = {
  'Content-Type': 'application/json',
  'Authorization': 'Token 26f9b124da0da589be592f666da4f45c3981f355'
}

response = requests.request("POST", url, headers=headers, data=payload)

# print(response.text)
df = response.json()
```

```python
def un_id(company_name):
    res = []
    for i in range(len(df['data'][company_name])):
        res.append(df['data'][company_name][i]['id'])
    return len(res),res 
```

```python
un_id('Samson Oil & Gas Limited_10k.pdf')
```
